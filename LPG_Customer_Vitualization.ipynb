```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly as px
```

```python
ds = pd.read_csv('LPGCust.csv')

ds1 = ds.groupby(['Region'],as_index=False).sum()

x = ds1['Region']
y = ds1['Total2020']
plt.bar(x,y,color='blue',width = 0.4)
plt.xlabel('Region in India')
plt.ylabel('Region in India')
plt.show()
```

```python
x = ds1['Region']
y = ds1['Total2020']
explode = (0,0.1,0)
fig1,ax = plt.subplots()
ax.pie(y, labels = x,autopct='%1.2f%%')
plt.show()
```

```python
def pieChart(x,y,title):
 colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
 fig = plt.figure()
 ax = fig.add_axes([0, 0, 1, 1])
 ax.axis('equal')
 plt.title(title)
 plt.pie(y, autopct='%1.2f%%', colors=colors, )
 centre_circle = plt.Circle((0, 0), 0.80, fc='white')
 fig = plt.gcf()
 fig.gca().add_artist(centre_circle)
 plt.legend(x, title='State',
           loc="upper right",
           )
 plt.show()
```

```python
import plotly.express as px
company = ['IOCL','HPCL','BPCL']
cust =  ds[company].sum()
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
plt.title('Companywise Share in LPG Connections')
plt.pie(cust, autopct='%1.2f%%', colors=colors, )
centre_circle = plt.Circle((0, 0), 0.80, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(company, title='Company',
           loc="upper right",
           )
plt.show()
```
